```python
'''
Model analyzing relationship between GDP, Urbanization, OilPricesGlobal and Oil_PerCapitalMWh
W/O Entity effects

'''
from linearmodels import PanelOLS
mod = PanelOLS.from_formula('PKT ~  GDP + Urbanization + OilPricesGlobal + Oil_PerCapitaMWh', data=etdata)
print(mod.fit())
```

```python
'''
Model analyzing relationship between GDP, Urbanization, OilPricesGlobal and Oil_PerCapitalMWh
'''
from linearmodels import PanelOLS
mod = PanelOLS.from_formula('PKT ~  GDP + Urbanization + OilPricesGlobal + Oil_PerCapitaMWh + EntityEffects', data=etdata)
print(mod.fit())
```

```python
'''
Model analyzing relationship between GDP, Urbanization, OilPricesGlobal and Oil_PerCapitalMWh
'''
from linearmodels import PanelOLS
mod = PanelOLS.from_formula('PKT ~  GDP + Urbanization + OilPricesGlobal + Oil_PerCapitaMWh + EntityEffects + TimeEffects', data=etdata)
print(mod.fit())
```

```python
'''
Model analyzing relationship between GDP, Urbanization, OilPricesGlobal and Oil_PerCapitalMWh
'''
from linearmodels import PanelOLS
mod = PanelOLS.from_formula('PKT ~  GDP + Urbanization + OilPricesGlobal + Oil_PerCapitaMWh + TimeEffects', data=etdata)
print(mod.fit())
```

```python
exog_vars = ['GDP','OilPricesGlobal']
exog = sm.add_constant(etdata[exog_vars])
```

```python
mod = RandomEffects(etdata.PKT, exog)
re_res = mod.fit()
print(re_res)
```
